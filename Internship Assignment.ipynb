```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/dev/student-data.csv")
```

```python
df.head()
```

```python
##Creating a new data frame for stroing labels to contain measure of fit according to student
dataframe = pd.DataFrame(index=np.arange(0,df.shape[0]),columns=['Fitness'])
```

```python
##Applying Conditions to attach label to each instance of dataset according to fitness of student for the assumed job 
##according to the data of the student


##For labelling each student according to values mentioned in the markdown above:
for ix in list(range(0,df.shape[0])):
    if(int(df.iloc[ix,-6])>=4):
        dataframe.iloc[ix] = 0
        if((int(df.iloc[ix,2])==15)or(int(df.iloc[ix,2])==16)):
            dataframe.iloc[ix] = 1
            if(int(df.iloc[ix,-4])<=2):
                dataframe.iloc[ix] = 2
for ix in list(range(0,df.shape[0])):
    if(int(df.iloc[ix,-6])==3):
        dataframe.iloc[ix] = 1
        if((int(df.iloc[ix,2])==17)or(int(df.iloc[ix,2])==18)):
            dataframe.iloc[ix] = 2
            if(int(df.iloc[ix,-4])==3):
                dataframe.iloc[ix] = 3
for ix in list(range(0,df.shape[0])):
    if(int(df.iloc[ix,-6])==2):
        dataframe.iloc[ix] = 1
        if((int(df.iloc[ix,2])==19)or(int(df.iloc[ix,2])==20)):
            dataframe.iloc[ix] = 3
            if(int(df.iloc[ix,-4])==4):
                dataframe.iloc[ix] = 3
for ix in list(range(0,df.shape[0])):
    if(int(df.iloc[ix,-6])==1):
        dataframe.iloc[ix] = 2
        if((int(df.iloc[ix,2])==21)or(int(df.iloc[ix,2])==22)):
            dataframe.iloc[ix] = 3
            if(int(df.iloc[ix,-4]) == 5):
                dataframe.iloc[ix] = 3
    if((int(df.iloc[ix,14])>0)and(df.iloc[ix,-1]=='yes')):
        dataframe.iloc[ix] = 1            
```

```python
#Concatenating label to main dataset
df = pd.concat([df,dataframe],axis=1)
```

```python
df_array = df.values
print(np.unique(df_array[:,-1],return_counts=True))
```

```python
df.head()
```

```python
## Removing not applicable values / incompatible type data
df = df.dropna()
##Label Encoding the string variables to integers
from sklearn.preprocessing import LabelEncoder
df.iloc[:,:-1] = df.iloc[:,:-1].apply(LabelEncoder().fit_transform)
```

```python
##Converting pandas DataFrame into Numpy Array 
data = df.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data[:,:-1]
Y = data[:,-1]
```

```python
##Splitting given data set into train, test set
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.25)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='newton-cg',multi_class='multinomial').fit(X_train,y_train)
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print(accuracy*100)
```

```python
from sklearn import tree
classify = tree.DecisionTreeClassifier()
classify.fit(X_train,y_train)
```

```python
y_p = classify.predict(X_test)
accurate = accuracy_score(y_test,y_p)
print(accurate*100)
```

```python
##Without removing redundant features, highest accuracy obtained with Gaussian Naive Bayes was 87%, so 
## we apply pca and select 3 components
from sklearn.decomposition import PCA
pca = PCA(n_components=12)
pca.fit(X)
X_pca = pca.transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_pca,Y,test_size = 0.25)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train)
y_prediction = gnb.predict(X_test)
print(accuracy_score(y_test,y_prediction)*100)
```
